```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

```python
y = np.array([0, 0, 0, 1])
```

```python
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10
```

```python
def step_function(z):
    return 1 if z >= 0 else 0
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}")

    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred

        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print(f"Input: {X[i]}, Target: {y[i]}, Predicted: {y_pred}, Error: {error}")

    print("Weights:", weights, "Bias:", bias)
    print("-" * 50)

print("Final Weights:", weights)
print("Final Bias:", bias)
```

```python
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
```

```python
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)
```

```python
grid_points = np.c_[xx.ravel(), yy.ravel()]
```

```python
Z = []
for point in grid_points:
    value = np.dot(point, weights) + bias
    Z.append(step_function(value))
```

```python
Z = np.array(Z).reshape(xx.shape)
```

```python
plt.figure(figsize=(7, 6))
plt.contourf(xx, yy, Z, alpha=0.3)

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i, 0], X[i, 1], marker='o', s=120, label='Class 0' if i == 0 else "")
    else:
        plt.scatter(X[i, 0], X[i, 1], marker='x', s=120, label='Class 1')

plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Perceptron Decision Region for AND Gate")
plt.legend()
plt.grid(True)
plt.show()
```
